# 00 — Setup smoke test

Verifies all dependencies import, the simulator instantiates, and a trivial VQE call completes. If this notebook runs top-to-bottom, the project is correctly installed.


```python
import sys
print('Python', sys.version)
import qiskit, qiskit_nature, qiskit_algorithms, qiskit_aer
print('qiskit', qiskit.__version__)
print('qiskit_nature', qiskit_nature.__version__)
print('qiskit_algorithms', qiskit_algorithms.__version__)
print('qiskit_aer', qiskit_aer.__version__)
import siam_vqe
print('siam_vqe', siam_vqe.__version__)
```

```python
from siam_vqe import hubbard_dimer, to_qubit_op, exact_diag
fop = hubbard_dimer(U=4.0, t=1.0)
pop = to_qubit_op(fop, scheme='parity_tapered', num_particles=(1, 1))
print(f'Hamiltonian: {pop.num_qubits} qubits, {len(pop)} Pauli terms')
ed = exact_diag(pop, k=1)
print(f'ED ground state energy: {ed.energies[0]:.6f}')
```

```python
from siam_vqe import efficient_su2_ansatz, run_vqe
circuit, x0 = efficient_su2_ansatz(num_qubits=pop.num_qubits, reps=2, seed=0)
result = run_vqe(pop, circuit, x0, optimizer='COBYLA', maxiter=200)
print(f'VQE energy: {result.energy:.6f}  ({len(result.history)} evals, {result.walltime_s:.2f}s)')
print(f'Gap to ED: {result.energy - ed.energies[0]:.2e}')
```

If everything above printed without errors and the VQE gap to ED is small (typically <1e-3 for this trivial problem), the project is installed correctly.
